Importing data

```python
import numpy as np
import pandas as pd
from scipy.io import loadmat

N_subjects = 50 ###### Needs to change based on data ######
data_dir = "blablabla"

## Import data from MATLAB file
roi_matrices = []
for i in range(N_subjects):
    
    temp_data = loadmat(f"data{i}.mat")

    roi_matrices.append(temp_data)
```

Compute Functional connectivity

```python
#Computes pearson correlation matrices for all subjects
fc_matrices = []
for roi in roi_matrices:
    for i in range(len(roi['data'])):
        fc_matrix = np.corrcoef(roi['data'][i])
        fc_matrices.append(fc_matrix)
```
